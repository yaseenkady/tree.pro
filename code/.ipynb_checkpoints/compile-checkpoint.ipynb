```python
import kraft
```

```python
na_pa = {}

na_se_li_ = {}

for pa in kraft.path.get_child("../raw/"):

    if pa[-3:] != ".md":

        continue

    na = pa.split("/")[-1][: -len(".md")]
    
    print(na)

    assert na == kraft.string.untitle(na)

    na_pa[na] = pa

    se_li_ = {
        ".": [],
        "<": [],
        "md<": [],
        ">": [],
        "md>": [],
    }

    se_ = []

    with open(pa) as io:

        for li in io:

            li = li.strip()

            if li == "":

                continue

            if li[0] == "#":

                se_.append(li[2])

                continue

            if li[0] == "(":

                assert li[-2] == "."

                continue

            se_li_[se_[-1]].append(li)

    assert all(se in se_ for se in [".", "<", ">"])

    assert any(0 < len(li_) for li_ in se_li_.values())

    na_se_li_[na] = se_li_

len(na_se_li_)
```

```python
for na, se_li_ in na_se_li_.items():

    for li in se_li_["<"]:

        assert kraft.string.untitle(li) not in na_se_li_, "{}: {}".format(na, li)
```

```python
import os
import shutil
```

```python
for na, se_li_ in na_se_li_.items():

    li_ = []

    for li in se_li_[">"]:

        limi = kraft.string.untitle(li)

        if limi in na_se_li_:

            se_li_["md>"].append(limi)

            na_se_li_[limi]["md<"].append(na)

        else:

            li_.append(li)

    se_li_[">"] = li_
```

```python
import pandas as pd
```

```python
di = "../compiled/"

if os.path.isdir(di):

    shutil.rmtree(di)

de = "\n\n"

for na, se_li_ in na_se_li_.items():

    pa = na_pa[na].replace("../raw/", di)

    kraft.path.make(pa, pr=False)

    with open(pa, "w") as io:

        ca_ = [li.upper() for li in se_li_["md<"]]

        fu_ = []

        fu_ += se_li_[">"]

        for li in se_li_["md>"]:

            fu_.append(li.upper())

            fu_ += na_se_li_[li][">"]

        for se, li_ in [
            ["Comment", se_li_["."]],
            ["Epidemiology", se_li_["<"]],
            ["Cause", ca_],
            ["Symptom and Sign", fu_],
        ]:

            io.write("# {}".format(se) + de)

            if len(li_) == 0:

                continue

            for li, n in pd.value_counts(li_).items():

                if 1 < n:

                    bbb = [na]

                    bbb += [
                        limd for limd in se_li_["md>"] if li in na_se_li_[limd][">"]
                    ]

                    print("{}:\n\t{}\n".format(li, "\n\t".join(bbb)))

            io.write(de.join(li_) + de)
```

```python
le_ = []

ri_ = []

for na, se_li_ in na_se_li_.items():

    le_ += se_li_["<"]

    ri_ += se_li_[">"]

for li_ in [le_, ri_]:

    print("-" * 80)

    li_n = pd.value_counts(li_)

    print(li_n)

pd.DataFrame(
    index=li_n.index.sort_values(), columns=["Diagnosis phage", "Physical exam phase"]
).to_csv("../feature.tsv", "\t")
```
